```python
from matplotlib import pyplot
from sklearn import datasets
import numpy as np

from L3.k_means import KMeans
```

```python
iris = datasets.load_iris()
data = np.array(iris.data)
data_partition = KMeans(k=3).fit(data[:, :2])
colormap = "rgb"
print(data_partition.iterations)
pyplot.scatter(data[:, 0], data[:, 1], 
               c=[colormap[i] for i in data_partition.groups], alpha=0.5)
pyplot.scatter(data_partition.group_centers[:, 0],
               data_partition.group_centers[:, 1], c="c")
pyplot.xlabel(iris.feature_names[0])
pyplot.ylabel(iris.feature_names[1])
pyplot.show()
```
