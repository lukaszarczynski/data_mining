```python
import numpy as np
import time

from L3.k_means import KMeans
from L3.zad1 import generate_data
```

```python
number = 1000000
dimension = 1000
distributions_number = 1000
probabilities = None
mi = np.array([10 * k * np.ones(dimension) for k in range(distributions_number)])
sigma = np.array([np.identity(dimension)] * distributions_number)

data = generate_data(number, dimension, distributions_number, probabilities, mi, sigma)
```

```python
t0 = time.time()
try:
    partition = KMeans(k=distributions_number)
    partition.fit(data, max_iterations=30)
except KeyboardInterrupt:
    pass
t1 = time.time()
print("Time: ", t1-t0)
```
