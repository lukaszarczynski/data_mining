```python
import numpy as np
import pandas
import matplotlib
from matplotlib import pyplot

from L3.k_means import KMeans
```

# Iris

```python
iris = pandas.read_csv('L3/data/iris.data', sep=',', header=None)
```

```python
iris = iris.ix[:, :3].values
```

```python
partition = KMeans(k=3)
partition.fit(iris)
```

```python
dimension = iris.shape[1]

for i in range(dimension):
    for j in range(dimension):
        pyplot.subplot(dimension, dimension, i*dimension+j+1)
        pyplot.scatter(iris[:, i], iris[:, j], c=partition.groups, marker='o', alpha=0.5)
        pyplot.scatter(partition.group_centers[:, i], partition.group_centers[:, j], c="r", marker='o')
pyplot.tight_layout()
pyplot.show()
```

# Wine

```python
wine = pandas.read_csv('L3/data/wine.data', sep=',', header=None)
wine = wine.loc[:,1:].values
```

```python
print(wine)
```

```python
partition = KMeans(k=3)
partition.fit(wine)
```

```python
dimension = wine.shape[1]

for i in range(dimension):
    pyplot.scatter(wine[:, i], wine[:, 12], 
           c=partition.groups, alpha=0.5)
    pyplot.scatter(partition.group_centers[:, i],
                   partition.group_centers[:, 12], c="k")
    pyplot.show()
print(partition.groups_sizes())
```

## Glass

```python
glass = pandas.read_csv('L3/data/glass.data', sep=',', header=None)
glass = glass.loc[:, [1,2,3,4,5,7,8,9]].values
```

```python
print(glass)
```

```python
partition = KMeans(k=7)
partition.fit(glass)
print(partition.groups_sizes())
dimension = glass.shape[1]
```

```python
colours = "r g b c m y".split() + ["xkcd:dark green", "xkcd:baby pink", 
                                     "xkcd:bright yellow", "xkcd:bright orange"]
for i in range(dimension):
    pyplot.scatter(glass[:, i], glass[:, i], 
           c=[colours[group] for group in partition.groups], alpha=0.7)
    pyplot.scatter(partition.group_centers[:, i],
                   partition.group_centers[:, i], c="k")
    pyplot.show()
```

```python
matplotlib.rcParams['figure.figsize'] = [24., 16.]

for i in range(4):
    for j in range(4):
        pyplot.subplot(dimension, dimension, i*dimension+j+1)
        pyplot.scatter(glass[:, i], glass[:, j], c=partition.groups, marker='o', alpha=0.5)
        pyplot.scatter(partition.group_centers[:, i], partition.group_centers[:, j], c="k", marker='o')
pyplot.show()

for i in range(4):
    for j in range(4, 8):
        pyplot.subplot(dimension, dimension, i*dimension+j+1)
        pyplot.scatter(glass[:, i], glass[:, j], c=partition.groups, marker='o', alpha=0.5)
        pyplot.scatter(partition.group_centers[:, i], partition.group_centers[:, j], c="k", marker='o')
pyplot.show()

for i in range(4, 8):
    for j in range(4):
        pyplot.subplot(dimension, dimension, i*dimension+j+1)
        pyplot.scatter(glass[:, i], glass[:, j], c=partition.groups, marker='o', alpha=0.5)
        pyplot.scatter(partition.group_centers[:, i], partition.group_centers[:, j], c="k", marker='o')
pyplot.show()

for i in range(4, 8):
    for j in range(4, 8):
        pyplot.subplot(dimension, dimension, i*dimension+j+1)
        pyplot.scatter(glass[:, i], glass[:, j], c=partition.groups, marker='o', alpha=0.5)
        pyplot.scatter(partition.group_centers[:, i], partition.group_centers[:, j], c="k", marker='o')
pyplot.show()
```
