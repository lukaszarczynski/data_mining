```python
from a_priori import APriori
import numpy as np
import time
```

```python
import pandas as pd
data = pd.read_excel('data/Online Retail.xlsx')
```

```python
data.head()
```

```python
data = data.drop(['StockCode', 'Quantity', 'InvoiceDate', 'UnitPrice', 'CustomerID', 'Country'], axis=1)
data.head()
```

```python
transactions = [list(data['Description'][x]) for x in data.groupby(['InvoiceNo'])['Description'].groups.values()]
transactions[:10]
```

```python
t0 = time.time()
a_priori = APriori(transactions, min_support=0.01, 
            min_confidence=0.75, min_leverage=0.01)
a_priori.find_frequent_sets()
print('Total time: {}'.format(time.time()-t0))
```

```python
import pickle
file = open("zad1.frequent_items", "wb")
pickle.dump(a_priori, file)
```
