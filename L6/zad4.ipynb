```python
from a_priori import APriori
import numpy as np
import time
```

```python
import pandas as pd
data = pd.read_csv('data/D01', encoding="Big5", 
                   sep=";", header=0, names=["date", "customer_id", 
                                             "age", "residence",
                                             "product_subclass", "product_id",
                                             "amount", "asset", "price"])
len(data)
# data.drop(0)
```

```python
data2 = pd.read_csv('data/D02', encoding="Big5", 
                    sep=";", header=0, names=["date", "customer_id", 
                                              "age", "residence",
                                              "product_subclass", "product_id",
                                              "amount", "asset", "price"])
data = data.append(data2, verify_integrity=True, ignore_index=True)
```

```python
data3 = pd.read_csv('data/D11', encoding="Big5", 
                    sep=";", header=0, names=["date", "customer_id", 
                                              "age", "residence",
                                              "product_subclass", "product_id",
                                              "amount", "asset", "price"])
data = data.append(data3, verify_integrity=True, ignore_index=True)
```

```python
data4 = pd.read_csv('data/D12', encoding="Big5", 
                    sep=";", header=0, names=["date", "customer_id", 
                                              "age", "residence",
                                              "product_subclass", "product_id",
                                              "amount", "asset", "price"])
data = data.append(data4, verify_integrity=True, ignore_index=True)
len(data)
```

```python
data.head()
```

```python
data = data.drop(["date", "age", "residence", "amount", "asset", "price"], axis=1)
data.head()
```

```python
data_subclass = data.drop(["product_id"], axis=1)
data_subclass.head()
```

```python
data_product_id = data.drop(["product_subclass"], axis=1)
data_product_id.head()
```

```python
data_product_id.groupby("customer_id").groups.values()
```

```python
data_product = data.drop(["customer_id"], axis=1)
data_product.head()
```

```python
transactions_product_id = [list(data_product_id['product_id'][x]) for x in 
                           data_product_id.groupby(['customer_id']).groups.values()]
transactions_product_id[:10]
```

```python
transactions_subclass = [list(data_subclass['product_subclass'][x]) for x in 
                           data_subclass.groupby(['customer_id']).groups.values()]
transactions_subclass[:10]
```

```python
t0 = time.time()
a_priori_product_id = APriori(transactions_product_id, min_support=0.01, 
            min_confidence=0.75, min_leverage=0.01)
a_priori_product_id.find_frequent_sets()

'Total time: {}'.format(time.time()-t0)
```

```python
t0 = time.time()
a_priori_subclass = APriori(transactions_subclass, min_support=0.01, 
            min_confidence=0.75, min_leverage=0.01)
a_priori_subclass.find_frequent_sets()

'Total time: {}'.format(time.time()-t0)
```

```python
import pickle
file = open("zad4_product_id.frequent_items", "wb")
pickle.dump(a_priori_product_id, file)
```
