```python
from matplotlib import pyplot
from L2.zad4 import polygon_point_clouds
```

```python
def plot(vertex_number, edge_length):
    points_clouds = polygon_point_clouds(vertex_number, edge_length)
    for points_cloud in points_clouds:
        center, points = points_cloud
        points_x, points_y = zip(*points)
        pyplot.scatter(points_x, points_y, marker=".", alpha=0.6)
    pyplot.title("K = {0}, d = {1}".format(vertex_number, edge_length))
    
    pyplot.show()
```

```python
for vertex_number in [7, 11, 23]:
    for edge_length in [5, 10, 15]:
        plot(vertex_number, edge_length)
```

```python
plot(4, 50)
```
