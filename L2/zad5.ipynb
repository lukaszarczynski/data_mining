```python
import numpy as np

from matplotlib import pyplot

from L2.zad5 import check_polygon_point_cloud_distances
```

```python
def plot(correct, incorrect):
    correct_x, correct_y = zip(*correct)
    incorrect_x, incorrect_y = [], []
    if len(incorrect) > 0:
        incorrect_x, incorrect_y = zip(*incorrect)
    pyplot.scatter(correct_x, correct_y, marker=".", alpha=0.6, color="green")
    pyplot.scatter(incorrect_x, incorrect_y, marker=".", alpha=0.6, color="red")
    incorrect_percent = len(incorrect) / (len(correct) + len(incorrect))
    pyplot.title("K = {0}, d = {1}, incorrect: {2:.2}%".format(vertex_number, edge_length, 
                                                            incorrect_percent)) 
                                                            
    pyplot.show()
```

```python
for vertex_number in [7, 11, 23]:
    for edge_length in [5, 10, 15]:
        correct, incorrect = check_polygon_point_cloud_distances(vertex_number, edge_length)
        plot(correct, incorrect)
```

```python
vertex_number = 7
for edge_length in np.arange(1, 4, 0.25):
    correct, incorrect = check_polygon_point_cloud_distances(vertex_number, edge_length)
    plot(correct, incorrect)
```
