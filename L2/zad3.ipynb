```python
import numpy as np
from matplotlib import pyplot
from math import sqrt
```

```python
def add_line(formula, x_range):
    x = np.array(x_range)  
    y = formula(x)
    pyplot.plot(x, y) 
```

```python
xs = np.arange(1, 11)
ys = np.array([10, 10, 11, 12, 18, 18, 19, 26, 19, 26])

points = list(zip(xs, ys))
```

```python
pyplot.scatter(xs, ys, marker='o', color="red", edgecolors="black")

pyplot.show()
```

```python
def draw(line_formula):
    pyplot.scatter(xs, ys, marker='o', color="red", edgecolors="black")
    pyplot.xlim(0, 11)
    pyplot.ylim(5, 27)
    add_line(line_formula, np.arange(0, 12))


line_formula = lambda x: 2*x+5
draw(line_formula)
pyplot.show()
```

```python
def sum_distances(line_formula, points):
    def general_equation(line_formula):
        c = -line_formula(0)
        b = 1
        a = -line_formula(1) - c
        return a, b, c
        
    def distance(line_formula, point):
        a, b, c = general_equation(line_formula)
        x, y = point
        dist = abs(a*x + b*y + c) / sqrt(a**2 + b**2)
        return dist
    
    distances = sum(distance(line_formula, point) for point in points)
    return distances
```

```python
print(sum_distances(line_formula, points))
draw(line_formula)
pyplot.savefig("./L2/zadanie3a.png")
pyplot.show()
```

```python
line_formula = lambda x: 2*x+6
print(sum_distances(line_formula, points))
draw(line_formula)
pyplot.savefig("./L2/zadanie3b.png")
pyplot.show()
```
