```python
import numpy as np
from sklearn import datasets
from matplotlib import pyplot

iris = datasets.load_iris()
```

```python
sepal_length = iris.data[:, iris.feature_names.index('sepal length (cm)')]
sepal_width = iris.data[:, iris.feature_names.index('sepal width (cm)')]


pyplot.title("Iris")
pyplot.xlabel('sepal length (cm)')
pyplot.ylabel('sepal width (cm)')

pyplot.scatter(sepal_length, sepal_width, marker='o', color="red", edgecolors="black")
pyplot.show()
```

```python
pyplot.title("Iris")
pyplot.xlabel('sepal length (cm)')
pyplot.ylabel('sepal width (cm)')

x_range = (3, 9)
y_range = (1, 5)
pyplot.xlim(x_range)
pyplot.ylim(y_range)

pyplot.scatter(sepal_length, sepal_width, marker='o', color="red", edgecolors="black")
pyplot.show()
```

```python
pyplot.title("Iris")
pyplot.xlabel('sepal length (cm)')
pyplot.ylabel('sepal width (cm)')
x_range = (3, 9)
y_range = (1, 5)
pyplot.xlim(x_range)
pyplot.ylim(y_range)

pyplot.xticks(np.arange(x_range[0], x_range[1]+1, 1))
pyplot.yticks(np.arange(y_range[0], y_range[1]+1, 1))

pyplot.scatter(sepal_length, sepal_width, marker='o', color="red", edgecolors="black")
pyplot.show()
```

```python
pyplot.title("Iris")
pyplot.xlabel('sepal length (cm)')
pyplot.ylabel('sepal width (cm)')
x_range = (3, 9)
y_range = (1, 5)
pyplot.xlim(x_range)
pyplot.ylim(y_range)

pyplot.xticks(np.arange(x_range[0], x_range[1]+1, 1))
pyplot.yticks(np.arange(y_range[0], y_range[1]+1, 1))

for target in set(iris.target):
    correct_target = (target == iris.target)
    pyplot.scatter(sepal_length[correct_target], sepal_width[correct_target], 
                   marker='o', edgecolors="black")

pyplot.savefig("./L2/zadanie1.png")
pyplot.show()
```
