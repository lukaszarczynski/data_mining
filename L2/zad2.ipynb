```python
import numpy as np
from sklearn import datasets
from matplotlib import pyplot

iris = datasets.load_iris()
```

```python
_sepal_length = iris.data[:, iris.feature_names.index('sepal length (cm)')]
_sepal_width = iris.data[:, iris.feature_names.index('sepal width (cm)')]
```

```python
iris.target_names
targets_to_plot = np.array([0, 2])
print(targets_to_plot)

sepal_length = np.array([])
sepal_width = np.array([])

setosa_sepal_length = np.append(sepal_length, _sepal_length[iris.target == 0])
setosa_sepal_width = np.append(sepal_width, _sepal_width[iris.target == 0])
virginica_sepal_length = np.append(sepal_length, _sepal_length[iris.target == 2])
virginica_sepal_width = np.append(sepal_width, _sepal_width[iris.target == 2])

x_range = (3, 9)
y_range = (1, 6)
```

```python
def plot(x_range, y_range):
    pyplot.title("Iris")
    pyplot.xlabel('sepal length (cm)')
    pyplot.ylabel('sepal width (cm)')
    pyplot.xlim(x_range)
    pyplot.ylim(y_range)
    pyplot.xticks(np.arange(x_range[0], x_range[1]+1, 1))
    pyplot.yticks(np.arange(y_range[0], y_range[1]+1, 1))

plot(x_range, y_range)
pyplot.scatter(setosa_sepal_length, setosa_sepal_width, 
               marker='o', color="red", edgecolors="black")
pyplot.scatter(virginica_sepal_length, virginica_sepal_width, 
               marker='o', color="red", edgecolors="black")

pyplot.show()
```

```python
def add_line(formula, x_range):
    x = np.array(x_range)  
    y = formula(x)
    pyplot.plot(x, y) 

plot(x_range, y_range)
line_func = lambda x: 2*x-8
add_line(line_func, x_range)

pyplot.scatter(setosa_sepal_length, setosa_sepal_width, 
               marker='o', color="red", edgecolors="black")
pyplot.scatter(virginica_sepal_length, virginica_sepal_width, 
               marker='o', color="red", edgecolors="black")

pyplot.show()
```

```python
def colour(line_func):
    colours = ["red", "green"]
    for setosa in zip(setosa_sepal_length, setosa_sepal_width):
        pyplot.scatter(setosa[0], setosa[1], 
                       marker='o', edgecolors="black",
                       color=colours[setosa[1]>line_func(setosa[0])])
    for virginica in zip(virginica_sepal_length, virginica_sepal_width):
        pyplot.scatter(virginica[0], virginica[1], 
                       marker='o', edgecolors="black",
                       color=colours[virginica[1]<line_func(virginica[0])])

plot(x_range, y_range)
line_func = lambda x: 2*x-8
add_line(line_func, x_range)
colour(line_func)

pyplot.savefig("./L2/zadanie2a.png")
pyplot.show()
```

```python
plot(x_range, y_range)
line_func = lambda x: 1.3*x-3.8
add_line(line_func, x_range)
colour(line_func)

pyplot.savefig("./L2/zadanie2b.png")
pyplot.show()
```
