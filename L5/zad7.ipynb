```python
% matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn import tree, ensemble, metrics

from L5.validation import test_training_indices, cross_validation, print_scores
```

```python
smartphone_targets = pd.read_csv("L5/data/y_train.txt", header=None)
len(smartphone_targets)
```

```python
smartphone_data = np.genfromtxt('L5/data/X_train.txt', dtype=np.float32)
len(smartphone_data)
```

```python
for max_leaf_nodes in range(10, 21, 2):
    classifier = tree.DecisionTreeClassifier(max_leaf_nodes=max_leaf_nodes)
    print_scores(classifier, smartphone_data, smartphone_targets, max_leaf_nodes)
```

```python
for min_samples_leaf in range(4, 21, 4):
    classifier = tree.DecisionTreeClassifier(min_samples_leaf=min_samples_leaf)
    print_scores(classifier, smartphone_data, smartphone_targets, min_samples_leaf)
```

```python
for max_depth in range(4, 11, 2):
    classifier = tree.DecisionTreeClassifier(max_depth=max_depth)
    print_scores(classifier, smartphone_data, smartphone_targets, max_depth)
```

```python
classifier = ensemble.ExtraTreesClassifier()
print_scores(classifier, smartphone_data, smartphone_targets.values.ravel(), min_samples_leaf)
```

```python
classifier = ensemble.RandomForestClassifier()
print_scores(classifier, smartphone_data, smartphone_targets.values.ravel(), min_samples_leaf)
```
