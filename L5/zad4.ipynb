```python
# % matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd

from sklearn import tree

from L5.validation import confusion_matrix, print_scores
```

```python
data = pd.read_csv("/home/lukasz/Pobrane/agaricus-lepiota.data", header=None)
data.head()
```

```python
data = data.apply(lambda x: pd.Categorical(x).codes)
data.head()
```

```python
mushroom_labels = data[0].values
mushroom_data = data.drop([0], axis = 1).values
```

```python
for max_leaf_nodes in range(10, 21, 2):
    classifier = tree.DecisionTreeClassifier(max_leaf_nodes=max_leaf_nodes)
    print_scores(classifier, mushroom_data, mushroom_labels, max_leaf_nodes)
```

```python
for max_leaf_nodes in range(10, 21, 2):
    classifier = tree.DecisionTreeClassifier(max_leaf_nodes=max_leaf_nodes, criterion="entropy")
    print_scores(classifier, mushroom_data, mushroom_labels, max_leaf_nodes)
```

```python
mushroom_tree = (tree.DecisionTreeClassifier(max_leaf_nodes=20, criterion="entropy")
        .fit(mushroom_data, mushroom_labels))
with open("L5/trees/mushroom.dot", "w") as f:
    tree.export_graphviz(mushroom_tree, out_file=f)
```

```python
confusion_matrix(
    tree.DecisionTreeClassifier(max_leaf_nodes=20, criterion="entropy"), 
    mushroom_data,
    mushroom_labels
)
```

```python
from sklearn import ensemble
```

```python
random_forest = ensemble.RandomForestClassifier()
random_forest.fit(mushroom_data, mushroom_labels)
print(random_forest.score(mushroom_data, mushroom_labels))
```

```python
import numpy as np
predictions = []
for decision_tree in random_forest.estimators_:
    predictions.append(decision_tree.predict(mushroom_data))
# print(next(iris_data.iterrows()))
# random_forest_iris.predict(iris_data)
predictions_array = np.array(predictions)
for i in range(len(mushroom_labels)):
    if len(set(predictions_array[:, i])) != 1:
        print(i, predictions_array[:, i])
```

```python
proximity_matrix = np.zeros((len(mushroom_labels), len(mushroom_labels)))
for tree_res_1 in predictions:
    for tree_res_2 in predictions:
        proximity_matrix += (np.array([tree_res_1]).T 
                             == np.array([tree_res_2])).astype(int)
proximity_matrix
```

```python
from sklearn.decomposition import PCA as sk_PCA

titanic_pca = sk_PCA().fit_transform(proximity_matrix)
# print(iris_pca.eigenvalues)
reduced_data = titanic_pca.T[:2].astype(float)
# print(reduced_data)
from matplotlib import pyplot

for target in set(mushroom_labels):
    correct_target = (target == mushroom_labels)
    pyplot.scatter(reduced_data[0][correct_target], reduced_data[1][correct_target], 
                   alpha=0.5)
pyplot.show()
```
