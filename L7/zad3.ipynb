```python
import numpy as np
from matplotlib import pyplot
from matplotlib.pyplot import imshow
from sklearn import decomposition
% matplotlib inline
```

```python
import scipy.io
mat = scipy.io.loadmat('L7/data/ReducedImagesForTraining.mat')
```

```python
print(mat["images"])
```

```python
images = np.array(mat["images"])
```

```python
print(images)
print(images.shape)
```

```python
print(images[0])
```

```python
print(np.min(images), np.max(images))
```

```python
image = images.T[0]
image_reshaped = image.reshape([60, 82]).T
```

```python
imshow(image_reshaped, cmap="gray")
```

```python
print(image_reshaped.shape)
```

```python
images_pca = decomposition.PCA().fit_transform(images)
```

```python
print(images_pca.shape)
```

```python
for i in range(20):
    eigenvector = images_pca.T[i].reshape([60, 82]).T
    imshow(eigenvector,
           cmap="gray")
    scipy.misc.imsave(f"L7/eigenvectors/face_{i}.png", eigenvector)
    pyplot.show()
```
