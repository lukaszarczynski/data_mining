```python
%matplotlib inline

import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

from sklearn.preprocessing import scale

path = '/pio/scratch/2/ED2015/APEX/APEX_OSD_Package_1.0/APEX_OSD_Package_1.0/APEX_OSD_V1_calibr_cube'
```

```python
from sklearn.metrics import silhouette_score
```

```python
data = np.fromfile(path, dtype=np.dtype('<i2'))
```

```python
X = data.reshape((285,-1)).T
X_scaled = scale(X.astype(np.float64), axis=0)
plt.rcParams['figure.figsize'] = (20.0, 16.0)
```

```python
plt.matshow(X.reshape((1500,1000,285))[:,:,0])
```

```python
k_means = KMeans(init='k-means++', n_clusters=3, n_init=1, verbose=2)
k_means.fit(X_scaled)
```

```python
C = k_means.labels_.reshape((1500,1000))
plt.rcParams['figure.figsize'] = (20.0, 16.0)
plt.matshow(C)
```

```python
k_means = KMeans(init='k-means++', n_clusters=5, n_init=10, verbose=2, n_jobs=-1)
k_means.fit(X_scaled)
```

```python
C = k_means.labels_.reshape((1500,1000))
plt.rcParams['figure.figsize'] = (20.0, 16.0)
plt.matshow(C)
```

```python
k_means2 = KMeans(init='k-means++', n_clusters=10, n_init=1, verbose=2, n_jobs=-1)
k_means2.fit(X_scaled)
```

```python
C = k_means2.labels_.reshape((1500,1000))
plt.matshow(C)
```

```python
k_means = KMeans(init='k-means++', n_clusters=15, n_init=2, verbose=2, n_jobs=-1)
k_means.fit(X_scaled)
C = k_means.labels_.reshape((1500,1000))
plt.matshow(C)
```
